```python
from skimage.color import rgb2gray
from skimage import io, img_as_ubyte, img_as_float
from skimage.filters import gaussian, laplace

import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
image_name = input("Caminho para imagem: ")
```

```python
image = io.imread(image_name)
image_grey = rgb2gray(image)
```

```python
io.imshow(image_grey)
```

### Aplicar filtro Gaussiano 

```python
kernel_size = float(input("Tamanho do kernel do Filtro Gaussiano: "))
im_gau = gaussian(image_grey, kernel_size, multichannel=False)
print("Resultado da aplicao do Filtro Gaussiano, com kernel = ", kernel_size)
io.imshow(im_gau)
```

# LoG da Imagem

```python
img_log = img_as_float(laplace(im_gau))
io.imshow(img_log)
```

# Filtro Laplaciano na imagem original

```python
lim = img_as_float(laplace(image_grey))
io.imshow(lim)
```

```python
kernel_size1 = float(input("Escolha o tamanho do primeiro kernel: "))
kernel_size2 = float(input("Escolha o tamanho do segundo kernel: "))
img_dog = gaussian(image_grey, kernel_size2, multichannel=False) - gaussian(image_grey, kernel_size1, multichannel=False)
io.imshow(img_dog)
```
